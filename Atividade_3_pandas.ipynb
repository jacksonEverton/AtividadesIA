```python
# Importar as bibliotecas necessÃ¡rias
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_csv('credit.csv')

# Exibir as primeiras linhas do dataset para entender a estrutura
df.head()
```

```python
# Agrupar por 'credit_history' e calcular a soma total de 'amount' e a idade mÃ©dia ('age')
grouped_credit_history = df.groupby('credit_history').agg(
    amount_sum=('amount', 'sum'),
    age_mean=('age', 'mean')
).reset_index()

# Exibir o resultado
grouped_credit_history
```

```python
# Calcular a mÃ©dia de 'amount' por categoria de 'purpose'
amount_mean_by_purpose = df.groupby('purpose')['amount'].transform('mean')

# Criar a coluna 'amount_ratio'
df['amount_ratio'] = df['amount'] / amount_mean_by_purpose

# Exibir o resultado
df[['purpose', 'amount', 'amount_ratio']].head()
```

```python
# FunÃ§Ã£o para classificar a idade
def classify_age(age):
    if age < 30:
        return 'Under 30'
    elif 30 <= age <= 50:
        return '30-50'
    else:
        return 'Over 50'

# Aplicar a funÃ§Ã£o para criar a coluna 'age_group'
df['age_group'] = df['age'].apply(classify_age)

# Exibir o resultado
df[['age', 'age_group']].head()
```

```python
# Agrupar por 'employment_duration' e 'housing' e calcular a mÃ©dia e o desvio padrÃ£o de 'amount'
grouped_employment_housing = df.groupby(['employment_duration', 'housing']).agg(
    amount_mean=('amount', 'mean'),
    amount_std=('amount', 'std')
).reset_index()

# Exibir o resultado
grouped_employment_housing
```

```python
# Agrupar por 'job' e calcular a idade mÃ­nima e mÃ¡xima
age_distribution_by_job = df.groupby('job').agg(
    age_min=('age', 'min'),
    age_max=('age', 'max')
).reset_index()

# Exibir o resultado
age_distribution_by_job
```

```python
# Calcular a mÃ©dia de 'amount' por 'credit_history'
amount_mean_by_credit_history = df.groupby('credit_history')['amount'].transform('mean')

# Criar a coluna 'amount_diff'
df['amount_diff'] = df['amount'] - amount_mean_by_credit_history

# Exibir o resultado
df[['credit_history', 'amount', 'amount_diff']].head()
```

```python
# Agrupar por 'checking_balance' e calcular a soma total de 'dependents' e a mÃ©dia de dependentes
grouped_checking_balance = df.groupby('checking_balance').agg(
    dependents_sum=('dependents', 'sum'),
    dependents_mean=('dependents', 'mean')
).reset_index()

# Exibir o resultado
grouped_checking_balance
```
